```python
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Stations, Hourly
import folium
import pandas as pd
```

```python
stations = Stations()
stations = stations.nearby(28.9, -95.6369)
station = stations.fetch(3)
station.to_json()
```

```python
# Create a map object
map = folium.Map(location=[station['latitude'].mean(), station['longitude'].mean()], zoom_start=10)

# Add markers to the map
for idx, row in station.iterrows():
    folium.Marker([row['latitude'], row['longitude']], popup=row['name']).add_to(map)
map
```

```python
station = stations.fetch(1)
start = datetime(2022, 1, 1)
end = datetime(2023, 2, 1)

# Get hourly data
data = Hourly(station, start, end)
data = data.fetch()


column_mapping_new = {
    'temp': 'air_temperature',
    'dwpt': 'dewpoint',
    'rhum': 'relative_humidity',
    'prcp': 'total_precipitation',
    'snow': 'snow_depth',
    'wdir': 'wind_direction',
    'wspd': 'average_wind_speed',
    'wpgt': 'wind_peak_gust',
    'pres': 'sea_level_air_pressure',
    'tsun': 'total_sunshine_duration',
    'coco': 'weather_condition_code'
}

# Rename columns using the new mapping
data.rename(columns=column_mapping_new, inplace=True)


```
