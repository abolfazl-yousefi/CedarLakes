```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.rcParams.update({'font.size': 12})

rasters_dict = {
    'Clipped(2001)_meter.tif': 'DEM 2001',
    'Clipped(2002)_meter.tif': 'DEM 2002',
    'Clipped(2006)_meter.tif': 'DEM 2006',
    'Clipped(2009)_meter.tif': 'DEM 2009',
    'Clipped(2016)_meter.tif': 'DEM 2016',
    'Clipped(2018)_meter.tif': 'DEM 2018',
    '(Clipped)2018_19_post_harvey_meter.tif': 'DEM 2018_19_post_harvey'
}

def build_combined_df(rasters_dict):
    combined_df = pd.DataFrame()
    for file_path, dem_year in rasters_dict.items():
        temp_df = pd.read_csv(f"Results/duneFeatures/Data/duneCharacteristics_{dem_year.replace(' ', '_')}.csv")
        temp_df['DEM Year'] = dem_year
        combined_df = pd.concat([combined_df, temp_df], ignore_index=True)
    return combined_df

def plot_ecdfs(combined_df, file_path):
    plt.figure(figsize=(10, 10))
    unique_years = combined_df['DEM Year'].unique()
    for i, column in enumerate(combined_df.columns[:-1], 1):
        plt.subplot(int(np.ceil((len(combined_df.columns) - 1) / 2)), 2, i)
        for year in unique_years:
            sns.ecdfplot(data=combined_df[combined_df['DEM Year'] == year], x=column)
        plt.tight_layout()
    plt.suptitle("Dune Characteristics - ECDF", fontsize=24)
    plt.subplots_adjust(top=0.92)
    # plt.savefig(f"{file_path}_ecdfs.png")
    # plt.close()

def plot_kdes(combined_df, file_path):
    plt.figure(figsize=(10, 10))
    unique_years = combined_df['DEM Year'].unique()
    for i, column in enumerate(combined_df.columns[:-1], 1):
        plt.subplot(int(np.ceil((len(combined_df.columns) - 1) / 2)), 2, i)
        for year in unique_years:
            sns.kdeplot(data=combined_df[combined_df['DEM Year'] == year], x=column, label=year)
        plt.legend()
        plt.tight_layout()
    plt.suptitle("Dune Characteristics - KDE", fontsize=24)
    plt.subplots_adjust(top=0.92)
    # plt.savefig(f"{file_path}_kdes.png")
    # plt.close()

def plot_boxplots(combined_df, file_path):
    plt.figure(figsize=(10, 10))
    for i, column in enumerate(combined_df.columns[:-1], 1):
        plt.subplot(int(np.ceil((len(combined_df.columns) - 1) / 2)), 2, i)
        sns.boxplot(x='DEM Year', y=column, data=combined_df)
        plt.tight_layout()
    plt.suptitle("Dune Characteristics - Boxplot", fontsize=24)
    plt.subplots_adjust(top=0.92)
    # plt.savefig(f"{file_path}_boxplots.png")
    # plt.close()

def process_and_plot(rasters_dict):
    combined_df = build_combined_df(rasters_dict)
    plot_boxplots(combined_df, "Results/combined")
    plot_kdes(combined_df, "Results/combined")
    plot_ecdfs(combined_df, "Results/combined")

process_and_plot(rasters_dict)
 
```
